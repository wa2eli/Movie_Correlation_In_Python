```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Now we need to read in the data
df = pd.read_csv(r'C:\Users\user\Desktop\LinkedIn data analysis\Projects\3 Correlation in Python\movies.csv')
```

```python
# Show data
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Create correct year col

df['CorrectedYear'] = df['released'].astype('str').str[9:14]
```

```python
df.head()
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
# Order our Data a little bit to see

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter Plot gross vs budget

plt.scatter(x=df["gross"], y=df["budget"])
plt.title("Budger vs Gross Earnings")
plt.xlabel("Gross Earnings")
plt.ylabel("Budget for Film")
plt.show()
```

```python
# Plot gross vs budget using Seaborn

sns.regplot(x="gross", y="budget", data=df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# Looking at correlation pearson, kendall, spearman
# Specify columns to calculate correlations between
columns_to_correlate = ['budget', 'gross', 'runtime', 'score', 'votes', 'year']
correlation_matrix = df[columns_to_correlate].corr(method ='pearson')
print(correlation_matrix)
```

```python
correlation_matrix = df[columns_to_correlate].corr(method ='kendall')
print(correlation_matrix)
```

```python
correlation_matrix = df[columns_to_correlate].corr(method ='spearman')
print(correlation_matrix)
```

```python
correlation_matrix = df[columns_to_correlate].corr()
print(correlation_matrix)
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```
